```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
file_path = '/Users/marufahmed/Work/Apurba/ocr_dataset_evaluation/levenshtein_result_ccw_word_latest_annotation.csv'
data = pd.read_csv(file_path)

# Assuming the column containing values between 0-100 is named 'your_column_name'
# Define the ranges for visualization (e.g., 0-10, 10-20, ..., 90-100)
ranges = list(range(0, 101, 5))

# Initialize counts for each range
counts = [sum((data['Adjusted_Score'] >= start) & (data['Adjusted_Score'] < start + 10)) for start in ranges[:-1]]

# Create labels for the ranges
labels = [f"{start}-{start + 4}" for start in ranges[:-1]]

# Plotting
plt.figure(figsize=(16, 12))
bars = plt.bar(labels, counts, color='skyblue')
plt.xlabel('Levenshtein Distance Score (higher is better, 100=perfect match)')
plt.ylabel('Count')
plt.title('Distribution of Levenshtein Values for CCW Word Dataset')
plt.xticks(rotation=45)

# Adding data labels
for bar, count in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.3, count, ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
file_path = '/Users/marufahmed/Work/Apurba/ocr_dataset_evaluation/levenshtein_result_tww_word_latest_annotation.csv'
data = pd.read_csv(file_path)

# Assuming the column containing values between 0-100 is named 'your_column_name'
# Define the ranges for visualization (e.g., 0-10, 10-20, ..., 90-100)
ranges = list(range(0, 101, 5))

# Initialize counts for each range
counts = [sum((data['Adjusted_Score'] >= start) & (data['Adjusted_Score'] < start + 10)) for start in ranges[:-1]]

# Create labels for the ranges
labels = [f"{start}-{start + 4}" for start in ranges[:-1]]

# Plotting
plt.figure(figsize=(12, 6))
bars = plt.bar(labels, counts, color='skyblue')
plt.xlabel('Levenshtein Distance Score (higher is better, 100=perfect match)')
plt.ylabel('Count (log scale)')
plt.title('Distribution of Levenshtein Values for TWW Word Dataset')
plt.xticks(rotation=45)

# Adding data labels
for bar, count in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.3, count, ha='center', va='bottom')

plt.yscale('log')  # Set y-axis to logarithmic scale

# Adding text information on top left
plt.text(0, 10000, 'Total: 200633 \nMismatch: 6442 \nError Rate: 3.21%', fontsize=12, color='black')


plt.tight_layout()
plt.show()
```
